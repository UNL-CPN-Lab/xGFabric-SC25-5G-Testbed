```python
import json
import matplotlib.pyplot as plt
import numpy as np
import os
import matplotlib.pyplot as plt
import numpy as np
import itertools
```

```python
def match_filter(path, filter_keywords):
    return all(keyword in path for keyword in filter_keywords)

def get_filenames(base_path, filter_keywords=None):
    files = []
    for root, _, filenames in os.walk(base_path):
        for filename in filenames:
            if filename.endswith('.json'):
                full_path = os.path.join(root, filename)
                if filter_keywords:
                    if not match_filter(full_path, filter_keywords):
                        continue
                files.append(full_path)
    return files

def key_for_dict(filepath, slicing=False):
    split1 = filepath.split("/")[-1]
    split2 = split2 = split1.split("_")[0]
    if slicing == True:
        split2 = split1.split("_")[1]
    return split2

def read_file(filepath):
    with open(filepath, 'r') as f:
        data = json.load(f)
    throughputs = [interval["sum"]["bits_per_second"] / 1_000_000
                   for interval in data.get("intervals", [])
                   if "sum" in interval and "bits_per_second" in interval["sum"]]
    return throughputs

def import_raw_data(files, slicing=False):
    raw_data = {}
    for filename in files:
        key = key_for_dict(filename, slicing)
        data = read_file(filename)
        if key not in raw_data:
            raw_data[key] = []
        raw_data[key].append(data)
    return raw_data

def get_stats(raw_data, metrics=["average", "std_dev"]):
    stats = {}
    for key, value in raw_data.items():
        flattened = [item for sublist in value for item in sublist]
        entry = {}
        if "average" in metrics:
            entry["average"] = np.mean(flattened)
        if "std_dev" in metrics:
            entry["std_dev"] = np.std(flattened)
        stats[key] = entry
    return stats

#get all the raw data, calculate mean/std, output as json, helper functions above
def process(filepath, filter_keywords=None, metrics=["average", "std_dev"], slicing=False):
    files = get_filenames(filepath, filter_keywords=filter_keywords)
    raw_data = import_raw_data(files, slicing)
    stats = get_stats(raw_data, metrics)
    return stats 
```

```python
##### Plot Single-UE Line #####

def plot_line_with_styles(label, dataset, max_bandwidth=None,
                          color='black', marker='o', linestyle='-'):
    filtered = {
        int(prb): metrics for prb, metrics in dataset.items()
        if max_bandwidth is None or int(prb) <= max_bandwidth
    }
    if not filtered:
        return

    x = sorted(filtered.keys())
    y = [filtered[prb]["average"] for prb in x]
    yerr = [filtered[prb]["std_dev"] for prb in x]

    plt.errorbar(x, y, yerr=yerr,
                 label=label,
                 color=color,
                 marker=marker,
                 linestyle=linestyle,
                 capsize=4,
                 linewidth=2)
    
def finalize_line_plot(filename="line.png"):
    plt.xlabel("Channel Bandwidth (MHz)", fontsize=28)
    plt.ylabel("Uplink Throughput (Mbps)", fontsize=28)
    plt.xticks(fontsize=28)
    plt.yticks(fontsize=28)
    plt.xlim(0, 55)
    plt.ylim(0, 100)
    plt.legend(fontsize=15, loc='upper left', framealpha=0.9)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.savefig(filename, dpi=300)
    plt.show()

def plot(line_data):
    keys = list(line_data.keys())

    # Style lists
    colors = [
        "#5DADE2", "#4682B4", "blue",
        "#CD5C5C", "#DC143C", "#F08080",
        "#8B0000", "#FF6347", "red", "#B22222"
    ]
    markers = ['o', '*', 's', 'D','D','s','*','o']
    linestyles = ['-', '--', '-.', ':']

    plt.figure(figsize=(12, 7))

    for index, key in enumerate(keys):
        plot_line_with_styles(
            label=key,
            dataset=line_data[key],
            max_bandwidth=50,
            color=colors[index % len(colors)],
            marker=markers[index % len(markers)],
            linestyle=linestyles[index % len(linestyles)]
        )

    finalize_line_plot("single-ue-uplink.pdf")
```

```python
##### Plot Multi-UE Line #####

def stacked_bar_plot(bar_data, filename="stacked_bar.png"):
    bandwidths = [5, 10, 15, 20, 30, 40, 50]
    bandwidth_labels = list(map(str, bandwidths))
    num_categories = len(bar_data)
    bar_width = 0.1
    x = np.arange(len(bandwidths))

    plt.figure(figsize=(14, 7))

    base_colors = {
        # 4G FDD
        "4G (FDD) Laptop": "#1E3F66",
        "4G (FDD) RPi": "#4682B4",
        "4G (FDD) Phone": "blue",
    
        # 5G FDD (lighter reds)
        "5G (FDD) Laptop": "#FA8072",  # Salmon
        "5G (FDD) RPi": "#F08080",     # Light Coral
        "5G (FDD) Phone": "#FF6347",   # Tomato
    
        # 5G TDD (darker reds)
        "5G (TDD) Laptop": "#C71515",  # Dark Red
        "5G (TDD) RPi": "red",     # red
        "5G (TDD) Phone": "#A52A2A",   # Brownish Red
    }
    
    # Stack colors
    blue_top = "#AED6F1"
    red_top = "#FFAFAF"  # Light pink
    hatches = ['/', 'x', '-', '\\', '.', '///', '||']
    
    offset = -((num_categories - 1) / 2) * bar_width

    for idx, (label, [d1, d2]) in enumerate(bar_data.items()):
        is_4g = "4G" in label
        base_color = base_colors.get(label, "#CCCCCC")
        if "4G" in label:
            top_color = blue_top
        elif "5G" in label:
            top_color = red_top

        y1 = [d1.get(str(bw), {}).get("average", 0) for bw in bandwidths]
        y2 = [d2.get(str(bw), {}).get("average", 0) for bw in bandwidths]
        bar_pos = x + offset + idx * bar_width

        # plt.bar(bar_pos, y1, width=bar_width, color=base_color, label=label)
        plt.bar(bar_pos, y1, width=bar_width, color=base_color, label=label, hatch=hatches[idx % len(hatches)], edgecolor='white')
        plt.bar(bar_pos, y2, width=bar_width, bottom=y1, color=top_color)

    # Unique legend
    handles, labels = plt.gca().get_legend_handles_labels()
    seen = set()
    unique = [(h, l) for h, l in zip(handles, labels) if l not in seen and not seen.add(l)]
    plt.legend(*zip(*unique), fontsize=22, loc='upper left')

    # Axes formatting
    plt.xticks(x, bandwidth_labels, fontsize=32)
    plt.yticks(fontsize=32)
    plt.xlabel("Channel Bandwidth (MHz)", fontsize=28)
    plt.ylabel("Uplink Throughput (Mbps)", fontsize=28)
    # plt.title("Two-User Uplink Throughput Across Devices", fontsize=18)
    plt.grid(axis="y", linestyle="--", alpha=0.5)
    plt.tight_layout()
    plt.savefig(filename, dpi=300)
    plt.show()
```

```python
##### Plot Slicing Line #####

def plot_single_line(label, dataset, color, marker='o', linestyle='-', order='ascending', x_base=None):
    prb_order = sorted(dataset.keys(), reverse=(order == 'descending'))
    y = [dataset[prb]['average'] for prb in prb_order]
    yerr = [dataset[prb]['std_dev'] for prb in prb_order]

    if x_base is None:
        x_base = list(range(len(prb_order)))

    plt.errorbar(
        x_base[:len(prb_order)], y, yerr=yerr,
        label=label,
        color=color,
        marker=marker,
        linestyle=linestyle,
        capsize=4,
        linewidth=2
    )
    return prb_order

def generate_merged_xticks(data, prb_orders):
    max_len = max(len(d) for d in data.values())
    merged_labels = []

    for i in range(max_len):
        labels_at_i = []
        for label, dataset in data.items():
            order = prb_orders.get(label, 'ascending')
            prb_order = sorted(dataset.keys(), reverse=(order == 'descending'))
            if i < len(prb_order):
                labels_at_i.append(str(prb_order[i]))
        merged_labels.append("\n".join(labels_at_i))
    
    return list(range(max_len)), merged_labels

def finalize_plot(x_base, xtick_labels, bw="0", filename="line.png"):
    plt.xticks(x_base, xtick_labels, fontsize=28)
    # plt.yticks()
    plt.xlabel("PRB Ratio (RPi1 | RPi2)", fontsize=28)
    plt.ylabel("Uplink Throughput (Mbps)", fontsize=28)
    # plt.title(f"Multi-UE Slicing Test on {bw} MHz TDD Cell", fontsize=18)
    plt.yticks(fontsize=28)

    ymin, ymax = plt.ylim()
    plt.ylim(0, ymax + 5)

    plt.legend(fontsize=22, loc='upper right', framealpha=0.9)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.savefig(filename, dpi=300)
    plt.show()
```
